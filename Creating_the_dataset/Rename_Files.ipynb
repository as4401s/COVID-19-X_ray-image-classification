```python
train_df = pd.read_csv('training_set.csv')
train_df.head()
```

```python
path_test = 'C:/Users/asarkar/Covid 19/data/test/'
test = os.listdir(path_test)
len(test)
```

```python
test_df = pd.read_csv('test_set.csv')
test_df.head()
```

```python
#making the directories
#please change path names according to where you want to save the files

if not os.path.exists('D:/Kaggle datasets/Covid_19_renamed_images'):
    try:
        os.mkdir('D:/Kaggle datasets/Covid_19_renamed_images')
        os.mkdir('D:/Kaggle datasets/Covid_19_renamed_images/train')
        os.mkdir('D:/Kaggle datasets/Covid_19_renamed_images/test')
    except OSError:
        print ("Creation of the directory %s failed")
    else:
        print ("Successfully created the directory %s ")
```

```python
image_size = 1024
```

```python
df1=train_df.drop(['Unnamed: 0','Sequence','Dataset'],axis=1)
df1.head()
```

```python
df2=test_df.drop(['Unnamed: 0','Sequence','Dataset'],axis=1)
df2.head()
```

```python
for i, item in df1.iterrows():

    try:
        img=sitk.ReadImage(os.path.join(path_train+item[0]))
        img_array=sitk.GetArrayFromImage(img)
    
        file_name = df1['File_name'][i]
        print(file_name)
        label = df1['Labels'][i]
        print(label)
        io.imsave('D:/Kaggle datasets/Covid_19_renamed_images/train/' + label +'_'+ str(i) + 'train.png', img_array) 
    
    except RuntimeError:
        
        pass
```

```python
for i, item in df2.iterrows():

    img=sitk.ReadImage(os.path.join(path_test+item[0]))
    img_array=sitk.GetArrayFromImage(img)
    
    file_name = df2['File_name'][i]
    print(file_name)
    label = df2['Labels'][i]
    print(label)
    io.imsave('D:/Kaggle datasets/Covid_19_renamed_images/test1/' + label +'_'+ str(i) + 'test.png', img_array) 
```
